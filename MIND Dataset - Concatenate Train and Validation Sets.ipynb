```
import pandas as pd
import numpy as np
import random
```

```
# behaviors.tsv: The click histories and impression logs of users

# Impression ID. The ID of an impression.
# User ID. The anonymous ID of a user.
# Time. The impression time with format “MM/DD/YYYY HH:MM:SS AM/PM”.
# History. The news click history (ID list of clicked news) of this user before this impression.
# Impressions. List of news displayed in this impression and user’s click behaviors on them (1 for click and 0 for non-click).

behaviors_train_small = pd.read_table('behaviors_train_small.tsv'
                                        , names=['impression_id','user_id','time','history','impressions']
                                        )

behaviors_train_small['set'] = 'train'

behaviors_validation_small = pd.read_table('behaviors_validation_small.tsv'
                                            , names=['impression_id','user_id','time','history','impressions']
                                            )

behaviors_validation_small['set'] = 'validation'

behaviors = pd.concat([behaviors_train_small, behaviors_validation_small])\
                .reset_index(drop=True)

print(behaviors_train_small.shape)
print(behaviors_validation_small.shape)
print(behaviors.shape)

behaviors.to_csv('behaviors.tsv', index=False, sep='\t')
```

```
# news.tsv: The information of news articles

news_train_small = pd.read_table('news_train_small.tsv'
                                    , names=['news_id'
                                            ,'category'
                                            ,'subcategory'
                                            ,'title'
                                            ,'abstract'
                                            ,'url'
                                            ,'title_entities'
                                            ,'abstract_entities']
                                    )

news_train_small['set'] = 'train'

news_validation_small = pd.read_table('news_validation_small.tsv'
                                        , names=['news_id'
                                                ,'category'
                                                ,'subcategory'
                                                ,'title'
                                                ,'abstract'
                                                ,'url'
                                                ,'title_entities'
                                                ,'abstract_entities']
                                    )

news_validation_small['set'] = 'validation'

news = pd.concat([news_train_small, news_validation_small])\
            .reset_index(drop=True)

print(news_train_small.shape)
print(news_validation_small.shape)
print(news.shape)

news.to_csv('news.tsv', index=False, sep='\t')
```
